### Cleaning the data from missing values

```python
#observing the dataset, it can be observed that there are negative values,
# this does not make sense with the data and needs to be addressed
df.drop(list(df.loc[df["sugars"]<1].index),inplace = True)
df.drop(list(df.loc[df["potass"]<1].index),inplace=True)
df.drop(list(df.loc[df["carbo"]<1].index),inplace=True)
```

```python
# The isna() will find any missing values in the rows. The sum() will add all the missing values.
df_miss = df.isna().sum()
df_miss
```

```python
# For mfr, it is an acronym for the brand. By using dictionary, will allow the user to view the name vs the acronym
cereal_mfr = {
  'A': 'American Home Food Products;',
  'G': 'General Mills',
  'K': 'Kelloggs',
  'N': 'Nabisco',
  'P': 'Post',
  'Q': 'Quaker Oats',
  'R': 'Ralston Purina ',
}
cereal_mfr
```

```python
cereal_rate = df['rating']
cereal_rate
```

```python
# Lets see any correalation by creating a correlation matrix
# the closer the number is to 1, presents a higher correlation. Numbers that go the opposite direction (-value)
# does not present a correlation
plt.figure(figsize=(20,10)) 
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
```

From the chart above, it can be observed that in correlating the data, fiber and protein have high correlation with ranking.

```python
# similar to a function we use lambda in a similar format to give it an expression for the given argument in this
#scenario the cereal_type
plt.figure(figsize=(10, 10))
df['type'].apply(
  lambda x: cereal_type[x],
).value_counts().plot.pie()
#need to find how to put in percentages
```

### Q3. Do Ingredients play a role in rating?

Upon looking in the correaltion matrix, it cam be observed that there is a higher correlation with ratings from fiber and protein correlation.

```python
# Lets group to values with rating to see if rating is consisted with an ingredients 
prorate = df.groupby(['protein', 'rating'])  
# Print the first value in each group
prorate.first()
```

```python
protein = df.groupby(["protein"]).mean().reset_index()
sns.displot(x="protein",y="rating",data=protein,hue="fiber")
```

```python
fiberrate = df.groupby(['fiber', 'rating']) 
# Print the first value in each group
fiberrate.first()
```

```python
fiber = df.groupby(["fiber"]).mean().reset_index()
sns.histplot(x="fiber",y="rating",data=fiber,hue="protein")
```
